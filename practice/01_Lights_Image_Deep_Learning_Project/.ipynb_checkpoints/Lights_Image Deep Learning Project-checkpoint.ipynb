```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# í›ˆë ¨ ë°ì´í„° ì´ë¯¸ì§€ ì „ì²˜ë¦¬ ë° í”½ì…€ ê°’ ì¡°ì • ê°ì²´ ì„ ì–¸
image_data_generator = ImageDataGenerator(rescale=1./255)
```

```python
# ë””ë ‰í† ë¦¬ì—ì„œ ì´ë¯¸ì§€ë¥¼ ê°€ì ¸ì™€ ë°°ì¹˜ë¡œ ë³€í™˜
generator = image_data_generator.flow_from_directory(root, target_size=(244, 244), batch_size=32, class_mode='categorical')
```

```python
import pandas as pd

# íŒŒì¼ ê²½ë¡œì™€ íƒ€ê²Ÿê°’ì„ ê°€ì§€ê³  ìƒˆë¡œìš´ ë°ì´í„° í”„ë ˆì„ ìƒì„±
l_df = pd.DataFrame({'file_paths': generator.filepaths, 'targets': generator.classes})
# ê²½ë¡œ ì¤‘ \\(ì—­ìŠ¬ë˜ì‹œ)ë¡œ ë˜ì–´ ìˆëŠ” ë¶€ë¶„ì„ /(ìŠ¬ë˜ì‹œ)ë¡œ ë³€ê²½
l_df.loc[:, 'file_paths'] = l_df.file_paths.apply(lambda x: x.replace('\\', '/'))
l_df
```

```python
from sklearn.model_selection import train_test_split

# í•™ìŠµ ë° ê²€ì¦ìš© ë°ì´í„° ì„¸íŠ¸ ë¶„ë¦¬
X_train, X_test, y_train, y_test = \
train_test_split(l_df.file_paths, l_df.targets, stratify=l_df.targets, test_size=0.2, random_state=124)
```

```python
import shutil

root = './datasets/light/'

for file_path in X_test:
    # ì´ë¯¸ì§€ íŒŒì¼ì˜ ì›ë³¸ ë””ë ‰í„°ë¦¬ ê²½ë¡œ ì €ì¥
    light_dir = file_path[len(root + 'original/'): file_path.rindex('/')]
    # ì´ë¯¸ì§€ íŒŒì¼ì„ ë³µì‚¬í•  í›ˆë ¨ìš© ë””ë ‰í„°ë¦¬ì˜ ê²½ë¡œ ì €ì¥
    destination = os.path.join(root, 'test/' + light_dir)

    # í›ˆë ¨ìš© ë””ë ‰í„°ë¦¬ê°€ ì—†ëŠ” ê²½ìš° í•´ë‹¹ ë””ë ‰í„°ë¦¬ë¥¼ ìƒì„±
    if not os.path.exists(destination):
        os.makedirs(destination)

    # ì´ë¯¸ì§€ íŒŒì¼ì„ í›ˆë ¨ìš© ë””ë ‰í„°ë¦¬ë¡œ ë³µì‚¬
    shutil.copy2(file_path, destination)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# ì…ë ¥ ë°ì´í„° í¬ê¸° ì§€ì •
IMAGE_SIZE = 244

# í›ˆë ¨, ê²€ì¦, í…ŒìŠ¤íŠ¸ ì´ë¯¸ì§€ í´ë” ê²½ë¡œ ì§€ì •
train_dir = './datasets/light/train'
validation_dir = './datasets/light/val'
test_dir = './datasets/light/test'

# ì´ë¯¸ì§€ì˜ í”½ì…€ ê°’ì„ 0ì—ì„œ 255 ì‚¬ì´ì—ì„œ 0ì—ì„œ 1 ì‚¬ì´ì˜ ê°’ìœ¼ë¡œ ì¡°ì •
train_data_generator = ImageDataGenerator(rescale=1./255)
validation_data_generator = ImageDataGenerator(rescale=1./255)
test_data_generator = ImageDataGenerator(rescale=1./255)

# ë””ë ‰í† ë¦¬ì—ì„œ ì´ë¯¸ì§€ë¥¼ ê°€ì ¸ì™€ ë°°ì¹˜ë¡œ ë³€í™˜
train_generator = train_data_generator.flow_from_directory(
    train_dir,
    target_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=32,
    class_mode='categorical'
)

validation_generator = validation_data_generator.flow_from_directory(
    validation_dir,
    target_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=32,
    class_mode='categorical'
)

test_generator = test_data_generator.flow_from_directory(
    test_dir,
    target_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=32,
    class_mode='categorical'
)

print(train_generator.class_indices)
print(validation_generator.class_indices)
print(test_generator.class_indices)
```

#### ğŸ“Š 1Cycle
- ë°ì´í„° ìˆ˜ê°€ ì ì–´ ë³µì¡ë„ë¥¼ ë‚®ì¶”ê¸° ìœ„í•´ ì¸µì„ ë§ì´ ìŒ“ì§€ ì•Šê³  ì§„í–‰.

```python
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dropout, Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model

# ì…ë ¥ ë°ì´í„° í¬ê¸° ì§€ì •
INPUT_SIZE = 244

# Input layer
input_tensor = Input(shape=(INPUT_SIZE, INPUT_SIZE, 3))

# Convolutional Layer 1
x = Conv2D(filters=32, kernel_size=3, strides=2, padding='same', activation='relu')(input_tensor)
x = Conv2D(filters=64, kernel_size=3, padding='same', activation='relu')(x)

# Flatten Layer: 2D í…ì„œë¥¼ 1Dë¡œ í‰íƒ„í™”
x = Flatten(name='classifierA00')(x)

# Dense Layer (ì€ë‹‰ì¸µ)
x = Dense(32, activation='relu', name='classifierAD01')(x)

# Dense Layer (ì¶œë ¥ì¸µ)
output = Dense(4, activation='softmax', name='output')(x)

# ì¸ê³µ ì‹ ê²½ë§ ìƒì„±
model = Model(inputs=input_tensor, outputs=output)
# ëª¨ë¸ êµ¬ì¡° í™•ì¸
model.summary()
```

```python
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import CategoricalCrossentropy

# ëª¨ë¸ ì»´íŒŒì¼
model.compile(optimizer=Adam(), loss=CategoricalCrossentropy(), metrics=['acc'])
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau, EarlyStopping

# weights ì €ì¥
mcp_cb = ModelCheckpoint(
    filepath="./callback_files/weights.{epoch:03d}-{val_loss:.4f}-{acc:.4f}.weights.h5",
    monitor='val_loss',
    save_best_only=False,
    save_weights_only=True,
    mode='min'
)

# ì¼ì • ê¸°ê°„ ë™ì•ˆ ì„±ëŠ¥ì´ ê°œì„ ë˜ì§€ ì•Šì„ ì‹œ í•™ìŠµë¥  ë™ì ìœ¼ë¡œ ê°ì†Œ
rlr_cb = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.1,
    patience=2,
    mode='min'
)

# ì¼ì • ê¸°ê°„ ë™ì•ˆ ì„±ëŠ¥ì´ ê°œì„ ë˜ì§€ ì•Šì„ ì‹œ í•™ìŠµ ì¡°ê¸° ì¤‘ë‹¨
ely_cb = EarlyStopping(
    monitor='val_loss',
    patience=3,
    mode='min'
)
```

```python
# í›ˆë ¨
history = model.fit(train_generator, 
                    batch_size=32, 
                    epochs=20,
                    validation_data=validation_generator,
                    callbacks=[mcp_cb, rlr_cb, ely_cb]
                   )
```

```python
# í…ŒìŠ¤íŠ¸ ë°ì´í„° ì„¸íŠ¸ë¡œ ëª¨ë¸ ì„±ëŠ¥ ê²€ì¦
model.evaluate(test_generator, batch_size=8)
```

```python
import matplotlib.pyplot as plt

# í•™ìŠµê³¼ì • ì§€í‘œë¥¼ ê·¸ë˜í”„í™”
plt.plot(history.history['acc'], label='train')
plt.plot(history.history['val_acc'], label='validation')
plt.legend()
```

#### ğŸ“Š 2Cycle
- ì´ë¯¸ì§€ ì „ì²˜ë¦¬ ì§„í–‰í•¨.
- ì´ë¯¸ì§€ì˜ ìƒí•˜, ì¢Œìš° ë°˜ì „ ë° ì‹ í˜¸ë“± ì´ë¯¸ì§€ ì´¬ì˜ì„ ì‹œê°„ì„ ê³ ë ¤í•˜ì—¬ ë°ê¸° ë° ëŒ€ë¹„ ì†ì„±ì„ ì¶”ê°€í•¨.

```python
import albumentations as A

# ì´ë¯¸ì§€ ì „ì²˜ë¦¬ í•¨ìˆ˜ ì„ ì–¸
def transform(image):
    aug = A.Compose([
        A.HorizontalFlip(p=0.5),
        A.VerticalFlip(p=0.5),
        A.RandomBrightnessContrast(brightness_limit=0.2, contrast_limit=0.2, p=0.5)
    ], p=0.5)

    return aug(image=image)['image']
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# í›ˆë ¨, ê²€ì¦, í…ŒìŠ¤íŠ¸ ì´ë¯¸ì§€ í´ë” ê²½ë¡œ ì§€ì •
train_dir = './datasets/light/train'

# ì…ë ¥ ë°ì´í„° í¬ê¸° ì§€ì •
IMAGE_SIZE = 244

# ì´ë¯¸ì§€ì˜ í”½ì…€ ê°’ì„ 0ì—ì„œ 255 ì‚¬ì´ì—ì„œ 0ì—ì„œ 1 ì‚¬ì´ì˜ ê°’ìœ¼ë¡œ ì¡°ì •
alb_image_data_generator = ImageDataGenerator(preprocessing_function=transform, rescale=1./255)

# ë””ë ‰í† ë¦¬ì—ì„œ ì´ë¯¸ì§€ë¥¼ ê°€ì ¸ì™€ ë°°ì¹˜ë¡œ ë³€í™˜
alb_train_generator = alb_image_data_generator.flow_from_directory(
    train_dir,
    target_size=(IMAGE_SIZE, IMAGE_SIZE),
    batch_size=32,
    class_mode='categorical'
)

print(alb_train_generator.class_indices)
```

```python
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dropout, Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model

# ì…ë ¥ ë°ì´í„° í¬ê¸° ì§€ì •
INPUT_SIZE = 244

# Input layer
input_tensor = Input(shape=(INPUT_SIZE, INPUT_SIZE, 3))

# Convolutional Layer 1
x = Conv2D(filters=32, kernel_size=3, strides=2, padding='same', activation='relu')(input_tensor)
x = Conv2D(filters=64, kernel_size=3, padding='same', activation='relu')(x)

# Flatten Layer: 2D í…ì„œë¥¼ 1Dë¡œ í‰íƒ„í™”
x = Flatten(name='classifierA00')(x)

# Dense Layer (ì€ë‹‰ì¸µ)
x = Dense(32, activation='relu', name='classifierAD01')(x)

# Dense Layer (ì¶œë ¥ì¸µ)
output = Dense(4, activation='softmax', name='output')(x)

# ì¸ê³µ ì‹ ê²½ë§ ìƒì„±
model = Model(inputs=input_tensor, outputs=output)
# ëª¨ë¸ êµ¬ì¡° í™•ì¸
model.summary()
```

```python
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import CategoricalCrossentropy

# ëª¨ë¸ ì»´íŒŒì¼
model.compile(optimizer=Adam(), loss=CategoricalCrossentropy(), metrics=['acc'])
```

```python
# í›ˆë ¨
history = model.fit(alb_train_generator, 
                    batch_size=32, 
                    epochs=10,
                    validation_data=validation_generator,
                    callbacks=[mcp_cb, rlr_cb, ely_cb]
                   )
```

```python
# í…ŒìŠ¤íŠ¸ ë°ì´í„° ì„¸íŠ¸ë¡œ ëª¨ë¸ ì„±ëŠ¥ ê²€ì¦
model.evaluate(test_generator, batch_size=8)
```

```python
import matplotlib.pyplot as plt

# í•™ìŠµê³¼ì • ì§€í‘œë¥¼ ê·¸ë˜í”„í™”
plt.plot(history.history['acc'], label='train')
plt.plot(history.history['val_acc'], label='validation')
plt.legend()
```

```python
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dropout, Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model

# ì…ë ¥ ë°ì´í„° í¬ê¸° ì§€ì •
INPUT_SIZE = 244

# Input layer
input_tensor = Input(shape=(INPUT_SIZE, INPUT_SIZE, 3))

# Convolutional Layer 1
x = Conv2D(filters=32, kernel_size=3, strides=2, padding='same', activation='relu')(input_tensor)
x = Conv2D(filters=64, kernel_size=3, strides=2, padding='same', activation='relu')(x)
x = Conv2D(filters=128, kernel_size=3, padding='same', activation='relu')(x)
x = Conv2D(filters=256, kernel_size=3, padding='same', activation='relu')(x)
x = Conv2D(filters=256, kernel_size=3, padding='same', activation='relu')(x)

# Flatten Layer: 2D í…ì„œë¥¼ 1Dë¡œ í‰íƒ„í™”
x = Flatten(name='classifierA00')(x)

# Dense Layer (ì€ë‹‰ì¸µ)
x = Dense(64, activation='relu', name='classifierAD01')(x)
x = Dense(32, activation='relu', name='classifierAD02')(x)

# Dense Layer (ì¶œë ¥ì¸µ)
output = Dense(4, activation='softmax', name='output')(x)

# ì¸ê³µ ì‹ ê²½ë§ ìƒì„±
model = Model(inputs=input_tensor, outputs=output)
# ëª¨ë¸ êµ¬ì¡° í™•ì¸
model.summary()
```

```python
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import CategoricalCrossentropy

# ëª¨ë¸ ì»´íŒŒì¼
model.compile(optimizer=Adam(), loss=CategoricalCrossentropy(), metrics=['acc'])
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau, EarlyStopping

# weights ì €ì¥
mcp_cb = ModelCheckpoint(
    filepath="./callback_files/weights.{epoch:03d}-{val_loss:.4f}-{acc:.4f}.weights.h5",
    monitor='val_loss',
    save_best_only=False,
    save_weights_only=True,
    mode='min'
)

# ì¼ì • ê¸°ê°„ ë™ì•ˆ ì„±ëŠ¥ì´ ê°œì„ ë˜ì§€ ì•Šì„ ì‹œ í•™ìŠµë¥  ë™ì ìœ¼ë¡œ ê°ì†Œ
rlr_cb = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.1,
    patience=2,
    mode='min'
)

# ì¼ì • ê¸°ê°„ ë™ì•ˆ ì„±ëŠ¥ì´ ê°œì„ ë˜ì§€ ì•Šì„ ì‹œ í•™ìŠµ ì¡°ê¸° ì¤‘ë‹¨
ely_cb = EarlyStopping(
    monitor='val_loss',
    patience=4,
    mode='min'
)
```

```python
# í›ˆë ¨
history = model.fit(alb_train_generator, 
                    batch_size=32, 
                    epochs=20,
                    validation_data=validation_generator,
                    callbacks=[mcp_cb, rlr_cb, ely_cb]
                   )
```

```python
# í…ŒìŠ¤íŠ¸ ë°ì´í„° ì„¸íŠ¸ë¡œ ëª¨ë¸ ì„±ëŠ¥ ê²€ì¦
model.evaluate(test_generator, batch_size=32)
```
